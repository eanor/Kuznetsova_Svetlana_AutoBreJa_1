```python
def tagging(smth):
  if smth == 'ADJF':
    smth = 'ADJ'
  elif smth == 'ADJS':
    smth = 'ADJ'
  elif smth == 'ADVB':
    smth = 'ADV'
  elif smth == 'GRND':
    smth = 'V'
  elif smth == 'INFN':
    smth = 'V'
  elif smth == 'NOUN':
    smth = 'N'
  elif smth == 'NPRO':
    smth = 'PRON'
  elif smth == 'NUMR':
    smth = 'NUM'
  elif smth == 'PREP':
    smth = 'PR'
  elif smth == 'PRTF':
    smth = 'V'
  elif smth == 'PRTS':
    smth = 'V'
  elif smth == 'ADP':
    smth = 'PR'
  elif smth == 'DET':
    smth = 'PRON'
  elif smth == 'PART':
    smth = 'PRCL'
  elif smth == 'SCONJ':
    smth = 'CONJ'
  elif smth == 'AUX':
    smth = 'V'
  elif smth == 'CCONJ':
    smth = 'CONJ'
  else:
    smth =smth
  return smth
```

```python
def ngrams(your_list, gram_1, gram_2):
  chonky_list1 = []
  for smth in range(len(your_list)-1):
    if (your_list[smth][0] == gram_1) or (your_list[smth][1] == gram_1):
      if (your_list[smth + 1][0] == gram_2) or (your_list[smth + 1][1] == gram_2):
        chonky_list = []
        chonky_list.append(your_list[smth][0])
        chonky_list.append(your_list[smth + 1][0])
        chonky_list1.append(chonky_list)
  return chonky_list1
```
